```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Data/0-Data_To_Prictice.csv', index_col=0)
x = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()    # n_estimators=200, learning_rate=0.05,  max_depth=4, subsample=0.8, colsample_bytree=0.8
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
score_accuracies = cross_val_score(estimator=classifier, X=x_train, y=y_train, cv=10, n_jobs=-1)
print(score_accuracies)
print(score_accuracies.mean())
```
