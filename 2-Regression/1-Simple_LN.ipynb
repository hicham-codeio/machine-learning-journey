```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
salary_data = pd.read_csv('Data/Salary_Data.csv', index_col=False)
salary_data.head()
salary_data.describe()
salary_data.info()
```

```python
x = salary_data.iloc[:,:-1].values
y = salary_data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=67)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
model.coef_
model.intercept_
```

```python
sns.regplot(x=predictions, y=y_test)
plt.title('Prediction vs True Values')
plt.xlabel('Predictions')
plt.ylabel('True Values')
```

```python
sns.regplot(x=x_train, y=y_train, color='red')
plt.title('Prediction vs True Values')
plt.xlabel('Predictions')
plt.ylabel('True Values')
```

```python
from sklearn import metrics
print(f"MAE = {metrics.mean_absolute_error(y_test, predictions)}")
print(f"MSE = {metrics.mean_squared_error(y_test, predictions)}")
print(f'RMSE = {metrics.root_mean_squared_error(y_test, predictions)}')
print(f"R2 = {metrics.r2_score(y_test, predictions)}")
```
