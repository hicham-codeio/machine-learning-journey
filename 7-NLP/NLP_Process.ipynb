```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

import nltk
from nltk import word_tokenize, WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re 
```

```python
dataset = pd.read_csv('Restaurant_Reviews.tsv', delimiter="\t", quoting=3)
dataset_cp = dataset.copy()
```

```python
dataset_cp.info()
```

```python
dataset_cp['Review'] = dataset_cp['Review'].apply(lambda sent : [word.lower() for word in sent.split()])
```

```python
en_words = stopwords.words('english')
en_words.remove('not')
dataset_cp['Review'] = dataset_cp["Review"].apply(lambda tokens : [token for token in tokens if token not in en_words ] )
```

```python
dataset_cp['Review'] = dataset_cp["Review"].apply(lambda tokens : [re.sub(r'[^\s\w]','',token) for token in tokens])
dataset_cp['Review'] = dataset_cp["Review"].apply(lambda tokens : ' '.join(tokens))
```

```python
dataset_cp['Review'] = dataset_cp["Review"].apply(lambda sent : word_tokenize(sent))
```

```python
#lemmatizer = WordNetLemmatizer()
#dataset_cp['Review'] = dataset_cp["Review"].apply(lambda tokens : [lemmatizer.lemmatize(token) for token in tokens])
```

```python
ps = PorterStemmer()
dataset_cp['Review'] = dataset_cp["Review"].apply(lambda tokens : [ps.stem(token) for token in tokens])
```

```python
X =  [','.join(map(str,token)) for token in dataset_cp['Review']]
y = dataset_cp.iloc[:,-1].values
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vec = CountVectorizer()
count_vec_trans = count_vec.fit_transform(X)
bow = pd.DataFrame(count_vec_trans.toarray(), columns=count_vec.get_feature_names_out())
```

```python
bow.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(bow, y, test_size=0.20, random_state=67)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix ,accuracy_score
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
